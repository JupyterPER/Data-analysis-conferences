```python
ds = pd.DataFrame(columns = ['scopus-id', 'first name', 'last name', 'full name', 'documents', 'citations', 'h-index', 'contribution'])
```

```python
# finding scopus data for all authors
notfound = []

for idx, row in dk.iterrows():
    # choose title and last name
    contribution = row['type']
    title = row['title'].replace('?','')
    fullname = row['authorslist'][0]
    authorlastname = fullname.split()[-1]
    authorfirstname = ' '.join(fullname.split()[:-1])
    try:
        # find in Scopus
        print(idx, ScopusDocQuery(title, authorlastname))
        s = ScopusSearch(ScopusDocQuery(title, authorlastname))
        doc = s.results[0]
        # save abstract
        dk.loc[idx, 'abstract'] = doc.description
        # for each author add scopus data to ds
        ids = list(map(int, doc.author_ids.split(';')))
        for scid in ids:
            author = AuthorRetrieval(scid)
            data = ScopusDataShort(author)
            data['contribution'] = contribution
            ds = ds.append(data, ignore_index=True)
    except TypeError:
        notfound+=[idx]
        print(f'{idx}, not found, {ScopusDocQuery(title, authorlastname)}')
        data = {'scopus-id': np.nan,
                'first name': authorfirstname,
                'last name': authorlastname,
                'full name': fullname,
                'documents': 0,
                'citations': 0,
                'h-index': 0}
        ds = ds.append(data, ignore_index=True)
    #break
```

```python
ds = ds.drop_duplicates('scopus-id').reset_index(drop = True)
ds
```

```python
cols = ['documents','citations', 'h-index']
ds[cols] = ds[cols].apply(pd.to_numeric, errors='coerce')
```
