## new columns

```python
da['last name'] = da['full name'].str.split().str[-1]
da['first name'] = da['full name'].apply(lambda name: ' '.join(name.split()[:-1]))
show_allrowscols(da)
```

```python
def initial(name):
    if name == '':
        return ''
    else:
        return name[0]
```

```python
da['ascii name'] = np_clean_str(da['first name'].apply(initial)+' '+da['last name'], case='lower', unicode=False)
show_allrowscols(da)
```

```python
da = da.drop_duplicates(subset='ascii name').reset_index(drop=True)
da = da.sort_values(['last name','first name']).reset_index(drop=True)
show_allrowscols(da)
```
