# for all conference contributions

```python
# dataframe for keywords
dkey = pd.DataFrame(columns = ['title', 'keywords'])
```

```python
# extracting title and keywords from all pdfs
notfound = []
for idx, pdfile in enumerate(pdfs):
    # open pdf file and the first page
    if 'organizers' not in pdfile and 'preface' not in pdfile:
        pdf = pdfp.open(pdfile)
        page1 = pdf.pages[1].extract_text()

        # page1 as dataframe
        dp1 = DF(page1.split('\n'), colname='line')
        dp1.line = dp1.line.apply(clean_str).replace('',np.nan)
        dp1 = dp1.dropna(how='all', axis='rows').reset_index(drop=True)

        # extract title
        data = dict()
        data['title'] = clean_str(dp1.line[0])
        if data['title'] == '':
            title = 'not found'
            notfound += [idx]
            print(f'{idx}. {pdfile}, title = not found')
        else:
            # extract keywords
            keywordline = dp1.line.str.contains('Keyword')
            if len(dp1[keywordline].index) == 1:
                data['keywords'] = dp1[keywordline].line.str.split(':').str[-1].apply(clean_str).values[0]
            else: 
                data['keywords'] = ''

            # add to keywords dataframe 
            dkey = dkey.append(data, ignore_index=True)
            print(f'{idx}. {pdfile}, title = {data["title"]}, keywords={data["keywords"]}')
```

### fuzzy matching

```python
# testing problematic titles
for idx, row in dkey.iterrows():
    title  = clean_str(row['title'], case='lower')
    keywords = row['keywords']
    titlecond = dk.titlelower.str.contains(title)
    if len(dk[titlecond].index) != 1:
           print(f'{idx}. problematic: {title} - {keywords}')
```

```python
# fuzzy matching text
from fuzzywuzzy import fuzz
ratio = lambda dktitle: fuzz.partial_ratio(title, dktitle)
```

```python
for idx, row in dkey.iterrows():
    title  = clean_str(row['title'], case='lower')
    keywords = row['keywords']
    titlecond = dk.titlelower.apply(ratio)>95
    dk.loc[titlecond, 'keywords'] = keywords
    if len(dk[titlecond].index) != 1:
           print(f'{idx}.{title} - {row["title"]}')
```

```python
dk = dk.drop(['titlelower'], axis='columns')
show_allrowscols(dk, fullcolwidth=True)
```
