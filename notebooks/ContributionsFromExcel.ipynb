# Contributions from Excel program

## parameters

```python
conference = 'DIDSCI+2022'
```

## basic processing

```python
# read xlsx program
dk = pd.read_excel(f'../{conference}-program.xlsx', header=None)
dk.head(3)
```

```python
# remove empty columns and rows
dk = dk.dropna(how='all', axis='columns')
dk = dk.dropna(how='all', axis='rows')
dk = dk.rename(columns={0:'authors',2:'title'})
dk
```

## invited papers

```python
dk['type'] = 'other'
dk.loc[dk.title.isna(),'type'] = 'invited'
```

```python
cond = dk.type == 'invited'
dk[cond]
```

```python
# split title and authors
dk.loc[cond,'title'] =dk.loc[cond,'authors'].str.split('\n').str[-1]
dk.loc[cond,'authors'] = dk.loc[cond,'authors'].apply(lambda item: ', '.join(item.split('\n')[:-1]))
dk[cond]
```

##  cleaning

```python
dk.authors = np_clean_str(dk.authors, replace=['.',''])
dk.title = np_clean_str(dk.title)
dk['authorslist'] = dk['authors'].str.replace(' and ',', ').str.split(', ')
```
